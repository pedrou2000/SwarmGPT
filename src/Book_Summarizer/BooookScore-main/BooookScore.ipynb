### Load Pickle Example File

```python
# Load pickle file to a variable
import pickle

pickle_file_path = 'example_all_books.pkl'
with open(pickle_file_path, 'rb') as f:
    all_books = pickle.load(f)

print(all_books.keys())
print(all_books)
```
